```python
#filtered_df = combined_df.query('Pos_x == "RB"') option with only one category
filtered_df = combined_df.loc[(combined_df.Pos_x == "RB") | (combined_df.Pos_x == "WR") | (combined_df.Pos_x == "TE")]
print(filtered_df)
```

```python
col = filtered_df.loc[: , 'Week1_PPR_points':'Week17_PPR_points'] 
filtered_df['mean'] = col.mean(axis=1)
filtered_df['SD'] = col.std(axis=1)
filtered_df['Sharpe'] = filtered_df['mean'] / filtered_df['SD']
filtered_df.dropna(subset=['SD'])
#nan_values = filtered_df[filtered_df['SD'].isna()]
```
